# 1)Read and Display the image

```python
import matplotlib.pyplot as plt
import cv2
```

```python
img=cv2.imread('img2.jpg')
```

```python
plt.imshow(img)
plt.show()
```

# 2)Draw a line in the image

```python
res=cv2.line(img,(0,0),(500,500),(205,100,25),10)
plt.imshow(res)
plt.axis('off')
plt.show()
```

# 3)Draw a circle at centre of the image

```python
imc=cv2.imread('img2.jpg')
resc=cv2.circle(imc,(250,250),150,(205,195,185),15)
```

```python
plt.imshow(resc)
plt.axis('on')
plt.show()
```

# 4)Draw a rectangle in the image

```python
imr=cv2.imread('img2.jpg')
start=(0,0)
stop=(500,500)
color=(150,150,100)
thick=35
```

```python
resr=cv2.rectangle(imr,start,stop,color,thick)
```

```python
plt.imshow(resr)
plt.show()
```

##### 5)Add text at top-left corner of the image

```python
imt = cv2.imread("img2.jpg")
text = "OpenCV Drawing"
position = (10, 50)
font = cv2.FONT_HERSHEY_COMPLEX_SMALL
font_scale = 1
color = (55, 155, 255) 
thickness = 4
rest = cv2.putText(imt, text, position, font, font_scale, color, thickness, cv2.LINE_AA)
```

```python
plt.imshow(rest)
plt.show()
```
